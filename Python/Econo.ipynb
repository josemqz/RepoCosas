```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
xi = np.linspace(0,10,10)
y1 = lambda x: x
y2 = lambda x: -x + 10
y3 = lambda x: -x + 7

plt.figure(figsize=(10,10))

plt.plot(xi,y1(xi), 'r', label="LM")
plt.plot(xi,y2(xi), 'c--', label="IS")
plt.plot(xi[:7],y3(xi[:7]), 'b', label="IS\'")

plt.xlabel("Y")
plt.ylabel("r")

plt.legend()
plt.show()
```
